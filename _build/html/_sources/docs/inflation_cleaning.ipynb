```python
import pandas as pd
import plotly.graph_objs as go
import plotly.express as px
import numpy as np
```

```python
inflation = pd.read_csv('inflatie.csv')
inflation.drop('Flag Codes', axis=1, inplace=True)
inflation.drop('FREQUENCY', axis=1, inplace=True)
```

```python
pd.DataFrame.head(inflation, n=5)
```

```python
countries = ['NLD', 'CHL', 'CHN', 'PRT', 
             'ZAF', 'CHE', 'CAN', 'SAU', 
             'NZL', 'JAP','IND']
time = ['2020', '2022']
subject = ['IDX2015']
sub = ['TOT']
inflatie = inflation[inflation['TIME'].isin(time)]
inflation2 = inflatie[inflatie['LOCATION'].isin(countries)]
inflation_def = inflation2[inflation2['MEASURE'].isin(subject)]
inflation_tot = inflation_def[inflation_def['SUBJECT'].isin(sub)]
pd.DataFrame.head(inflation_tot, n=20)
```
